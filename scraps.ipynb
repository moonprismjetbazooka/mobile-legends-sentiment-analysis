```python
import os
import pandas as pd
from google_play_scraper import Sort, reviews
```

```python
results = reviews(
    'com.mobile.legends',
    lang='id',
    country='id',
    count=60000,
    sort=Sort.MOST_RELEVANT
)
```

```python
review_df = pd.DataFrame(results[0])
review_df.tail()
```

```python
data_folder_path = os.path.join(os.getcwd(),'data')

if not os.path.exists(data_folder_path):
    print('Creating Directory...')
    os.makedirs(data_folder_path)

print('Writing CSV File...')
review_df.to_csv(os.path.join(data_folder_path,r'mobile_legends_review.csv'))
```
